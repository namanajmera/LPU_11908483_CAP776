```python
import matplotlib.pyplot as pltbar

x=ageGroupDetails['AgeGroup']
y=ageGroupDetails['TotalCases']
pltbar.xlabel('Age Groups')
pltbar.ylabel('Total Cases')
pltbar.title("Total Corona Cases according to Age Groups")
pltbar.bar(x,y)
```

```python
# group countries by continents and apply sum() function 
HospitalBedsIndiaState=HospitalBedsIndia.groupby('State/UT', axis=0).sum()

# note: the output of the groupby method is a `groupby' object. 
# we can not use it further until we apply a function (eg .sum())
print(type(HospitalBedsIndia.groupby('State/UT', axis=0)))

HospitalBedsIndiaState
```

```python
import matplotlib.pyplot as pltpie

HospitalBedsIndiaState['Total_Beds'].plot(kind='pie',
                            figsize=(15,18),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90Â°
                            shadow=True, # add shadow
                            
                            )

pltpie.title('Total Beds in Different State and UT')
pltpie.axis('equal') # Sets the pie chart to look like a circle.

pltpie.show()
```

```python
StatewiseTestingDetailsState=StatewiseTestingDetails.groupby('State',axis=0).sum()

print(type(StatewiseTestingDetails.groupby('State', axis=0)))

StatewiseTestingDetailsState
```

```python
import matplotlib.pyplot as plt

StatewiseTestingDetailsState['Positive'].plot(kind='pie',
                            figsize=(19,8),
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90Â°
                            pctdistance=1.12,
                            labels=None,
                            shadow=True,       # add shadow      
                            )

plt.title('Total Positive Cases according to State')
plt.legend(labels=StatewiseTestingDetailsState.index, loc='upper left') 
plt.axis('equal') # Sets the pie chart to look like a circle.

plt.show()
```
